```python
import pandas as pd

df = pd.read_csv('data/data.csv')
df = df[df['Inc_state_truth'].notna()]
df = df.reset_index(drop=True)
to_drop = ['Inc_state_Regex', 'Inc_state_Hybrid', 'Inc_state_LLM', 'Inc_state_Match', 'Inc_year_Regex', 'Inc_year_Hybrid', 'Inc_year_LLM', 'Inc_year_Match']
df.drop(columns=to_drop, inplace=True)
df = df[['filename', 'cik', 'year', 'Inc_state_truth']]
df.to_csv('data/cleaned_data.csv', index=False)
```
